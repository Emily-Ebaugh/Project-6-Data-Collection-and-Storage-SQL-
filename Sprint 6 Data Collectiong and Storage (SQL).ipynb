## Hypothesis Testing
In addition to identifying popular drop-off neighborhoods, we will also test the hypothesis:

"The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

### Setting the Significance Level
For this analysis, we will set the significance level (alpha) at 0.05. This is a common threshold in hypothesis testing, indicating that we are willing to accept a 5% chance of incorrectly rejecting the null hypothesis (Type I error).


### Formulation of Hypotheses
Null Hypothesis (H0): The average duration of rides from the Loop to O'Hare International Airport does not change on rainy Saturdays. In statistical terms, this can be expressed as: [ H_0: \mu_{\text{rainy}} = \mu_{\text{non-rainy}} ] where (\mu_{\text{rainy}}) is the average duration of rides on rainy Saturdays, and (\mu_{\text{non-rainy}}) is the average duration of rides on non-rainy Saturdays.

### Alternative Hypothesis (H1):
The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays. This can be expressed as: [ H_1: \mu_{\text{rainy}} \neq \mu_{\text{non-rainy}} ] This is a two-tailed test, as we are interested in any change (increase or decrease) in ride duration.


### Criterion for Testing the Hypotheses
To test the hypotheses, we will use the following statistical approach:

Data Collection: We will collect the ride duration data for rides from the Loop to O'Hare on Saturdays, categorizing them into two groups based on weather conditions:

Rides on rainy Saturdays (where the weather condition is classified as "Bad").
Rides on non-rainy Saturdays (where the weather condition is classified as "Good").
Statistical Test: We will perform an independent two-sample t-test to compare the means of the two groups (rainy vs. non-rainy). The t-test is appropriate here because we are comparing the means of two independent samples.

Calculation of Test Statistic: The t-test will yield a t-statistic, which we will compare against the critical t-value from the t-distribution table based on our chosen alpha level (0.05) and the degrees of freedom.


### Decision Rule:

If the p-value obtained from the t-test is less than the significance level (0.05), we will reject the null hypothesis (H0).
If the p-value is greater than or equal to 0.05, we will fail to reject the null hypothesis.
Conclusion: Based on the results of the t-test, we will draw conclusions about whether there is a statistically significant difference in the average duration of rides on rainy Saturdays compared to non-rainy Saturdays.

## 2: Set the Significance Level
Significance Level (α): We will set the significance level at ( \alpha = 0.05 ). This is a common choice in hypothesis testing, indicating that we are willing to accept a 5% chance of incorrectly rejecting the null hypothesis (Type I error).

## 3: Collect and Prepare the Data
We will use the provided dataset (/datasets/project_sql_result_07.csv), (/datasets/project_sql_result_04.csv) and '/datasets/project_sql_result_01.csv')  to extract the relevant data for rides on rainy and non-rainy Saturdays. Rainy and non-rainy days are going to be seperated by Good and Bad days listed in the dataframe provided. 

## 4. Prepare Data for Hypothesis Testing: 
The code converts the start_ts column to datetime, extracts the day of the week, and identifies rainy conditions.



## 5. Filter Data: The DataFrame is filtered to separate rainy and non-rainy Saturdays.
Perform T-Test: An independent two-sample t-test is performed to compare the average durations of rides on rainy and non-rainy Saturdays.
Conclusion: The results of the t-test are printed, and a conclusion is drawn based on the p-value compared to the significance level.

## 6. Conclusion: The results of the t-test are printed, and a conclusion is drawn based on the p-value compared to the significance level.

````python

# Convert 'start_ts' to datetime in the weather_df
weather_df['start_ts'] = pd.to_datetime(weather_df['start_ts'])

# Extract the day of the week
weather_df['Day of Week'] = weather_df['start_ts'].dt.day_name()

# Define rainy conditions based on the dataset's weather conditions
rainy_conditions = ['Bad']

# Create a column to identify rainy days
weather_df['Is Rainy'] = weather_df['weather_conditions'].isin(rainy_conditions)

# Filter for Saturdays
saturdays_df = weather_df[weather_df['Day of Week'] == 'Saturday']

# Separate rainy and non-rainy Saturdays
rainy_saturdays = saturdays_df[saturdays_df['Is Rainy']]
non_rainy_saturdays = saturdays_df[~saturdays_df['Is Rainy']]

# Calculate average durations
average_durations = {
    'Rainy Saturdays': rainy_saturdays['duration_seconds'].mean(),
    'Non-Rainy Saturdays': non_rainy_saturdays['duration_seconds'].mean()
}

# Print average durations
print("Average Duration of Rides:")
print(f"Rainy Saturdays: {average_durations['Rainy Saturdays']} seconds")
print(f"Non-Rainy Saturdays: {average_durations['Non-Rainy Saturdays']} seconds")

# Set the significance level
alpha = 0.05

# Perform the t-test if both groups have data
if len(rainy_saturdays) > 0 and len(non_rainy_saturdays) > 0:
    t_stat, p_value = stats.ttest_ind(
        rainy_saturdays['duration_seconds'].dropna(),
        non_rainy_saturdays['duration_seconds'].dropna(),
        equal_var=False  # Use Welch's t-test if variances are not equal
    )

    # Print the results
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    # Conclusion based on the p-value
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in average duration on rainy Saturdays.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in average duration on rainy Saturdays.")
else:
    print("Insufficient data to perform t-test.")
````
